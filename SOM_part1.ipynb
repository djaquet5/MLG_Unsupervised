## Introduction

In this laboratory, we will work with a different type of problem called unsupervised learning.
When we talk about unsupervised learning, it means that we are trying to find **hidden structure** in unlabeled data. Therefore, no simple measure exists to evaluate a potential solution.


We will focus on a commonly used algorithm to solve this kind of problem called "Self Organizing Maps".
As you saw in the theoretical part of the course, the main advantage of this algorithm is its topological ordering, which means that the maps can be visualized as **elastic grids**. Therefore they can be easily plotted.

### Definition of the problem

In this problem, we will apply the SOM algorithm as follows:
    
1. Initialize SOM with its parameters (such as size of the map, neighborood size, learning rate...)
2. Sequentially present a random sample animal for the map to learn
3. Repeat step 2 until the whole dataset was learned t times

After the SOM has been "trained", we can then visualize it by plotting its U-Matrix (also known as Unified Distance Matrix) which shows the neurons and the distance with their neighbors.
To construct a U-Matrix, we need to:

1. Create a cell for each neuron
2. Create a cell for each 1-1 connection between two neurons
3. Assign these new cells the distance between the two neurons they connect (higher is further apart)

In the end, if we have a very simple SOM (2x2):

    Neuron(0) - Neuron(1)
       |           |
    Neuron(2) - Neuron(3)

The corresponding U-Matrix (2\*2 -1=3, 2\*2 -1=3) will be:

    Neuron(0) - Dist(0,1) - Neuron(1)
       |           |           |
    Dist(0,2) -  ?????    - Dist(1,3)
       |           |           |
    Neuron(2) - Dist(2,3) - Neuron(3)
    

We are only going to plot the distances.

Let us present the animal dataset 10 times to a newly generated kohonen map. We can then compute its U-matrix and plot the result.

### First tests with SOM and U-Matrix

```python
kmap = KohonenUtils.KohonenMap(20,
                               20,
                               size_vector=len(animals.columns))
n_iter = 10
learning_rate = KohonenUtils.ExponentialTimeseries(1, 0.2, n_iter * len(animals.index))
neighborhood_size = KohonenUtils.ExponentialTimeseries(4./3 * len(animals.index), 1, n_iter * len(animals.index))

kmap.train(animals.values, np.array(animals.index), n_iter, learning_rate, neighborhood_size)

kmap.plot_umatrix(plot_names=True, plot_empty=True, fontsize=20)
```

## Exercice 3 : Wine database clustering with K-Means

Now, use the K-means algorithm to separate the three types of wine in the Wine dataset (K = 3).

```python
from sklearn.datasets import load_wine

data = load_wine()
features = data.data
classes = data.target
classes_names = data.target_names
```

```python
kmeans = KMeans(n_clusters=3,n_init=10, init='k-means++').fit(features)
```

```python
print("Result\n", kmeans.labels_)
print("\n", "Classes\n", classes)
```

<p style="background-color:#006600; color:#fff;padding:5px; font-weight:bold">Q: Observe the observations that are grouped together by K-Means. </p>

<p>Après plusieurs compilations, on peut s'apercevoir que les numéros des prototypes changent régulierement. Ce changement est dû au fait que les premiers prototypes sont initialisés de manière aléatoire. Cependant, les groupes restent en grande partie identique d'une itération à l'autre.</p>



<p style="background-color:#006600; color:#fff;padding:5px; font-weight:bold">Q: Count the number of "bottles" that are correctly grouped. What is the accuracy of this unsupervised classification?</p>

```python
def get_nb_elements(array, start, length):
    # src : https://stackoverflow.com/questions/28663856/how-to-count-the-occurrence-of-certain-item-in-an-ndarray-in-python
    _, counts = np.unique(array[start:length], return_counts=True)
    return counts
    

def get_accuracy(predictions):
    classes_length = get_nb_elements(classes, 0, len(classes))
    nb_first = classes_length[0]
    nb_second = classes_length[1]
    nb_third = classes_length[2]
    
    # Count first class
    first_class = get_nb_elements(predictions, 0, nb_first)
    accuracy_first = np.amax(first_class)/np.sum(first_class)
    
    # Count second class
    second_class = get_nb_elements(predictions, nb_first, nb_first+nb_second)
    accuracy_second = np.amax(second_class)/np.sum(second_class)
    
    # Count third class
    third_class = get_nb_elements(predictions, nb_first+nb_second, nb_first+nb_second+nb_third)
    accuracy_third = np.amax(third_class)/np.sum(third_class)
    
    return (accuracy_first+accuracy_second+accuracy_third)/3
    
    
def calculate_accuracy(features, n=10):
    accuracy = 0
    for i in range(0, 10):
        kmeans = KMeans(n_clusters=3,n_init=10, init='k-means++').fit(features)
        accuracy += get_accuracy(kmeans.labels_)
    
    print("Last execution:\n", kmeans.labels_)
    print(f"\nAccuracy is {accuracy/n:.2%}")
    
    
calculate_accuracy(features)
```

<p style="background-color:#006600; color:#fff;padding:5px; font-weight:bold">Q: Try to improve the performance of the classification. Does normalizing the data increases the accuracy? Does selecting a reduced number of features improves the accuracy? Why?. </p>

<p>Le fait de normaliser les données permet de rendre compatibles des données qui ne le seraient pas forcément. Par exemple, si une feature était le degré d'alcool et une autre le poid de la bouteille, ils ne sont pas réellement comparables. Il faut donc les normaliser pour minimiser la variance.</p>

```python
# Normalize data per columns between min and max
def scale(array, min, max):
    nom = (array-array.min(axis=0))*(max-min)
    denom = array.max(axis=0) - array.min(axis=0)
    denom[denom==0] = 1
    return min + nom/denom

scaled_features = scale(features, 0, 1)
calculate_accuracy(scaled_features)
```

<p> Le fait de réduire les features peut impacter le résultat dans les deux sens. Si certaines features sont trop similaires entre les classes, le fait de les supprimer aura un impact positif. Dans la même idée, il est possible de supprimer des features permettant de distinguer des types de classe plus facilement, on aura donc un résultat moins bon.</p>

```python
small_features = np.delete(features, slice(0, 10), 1)
calculate_accuracy(small_features)
```

<p>Ci-dessous, nous avons donc les deux principes ensembles, on aura donc une variance minimisée et aussi des données plus précise pour séparer les données. On ne doit pas oublier l'impacte que chaque feature a sur le résulat final.</p>

```python
small_scaled_features = scale(np.delete(features, slice(0, 10), 1), 0, 1)
calculate_accuracy(small_scaled_features)
```

<p> On a le même cas que le précédent sauf qu'ici la feature supprimée a un plus gros impact la séparation des données par l'algorithme `k-means++`. On constate donc que l'accuracy est moins bonne.</p>

```python
firsts_features = np.delete(features, 12, 1)
calculate_accuracy(firsts_features)
```
